# Face Detection:

```python
#Loading the image to be tested
test_image = cv2.imread('face_1.jpg')

#Converting to grayscale
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)

# Displaying the grayscale image
plt.imshow(test_image_gray, cmap='gray')
# Since we know that OpenCV loads an image in BGR format, so we need to convert it into RBG format to be 
# able to display its true colors. Let us write a small function for that.
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier('/Users/imanory/PycharmProjects/untitled/venv/lib/python3.7/site-packages/cv2/data/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('/usr/local/Cellar/opencv/3.4.0_1/share/OpenCV/haarcascades/haarcascade_eye.xml')

faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, 1.2, 5);

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
        
#convert image to RGB and show image
plt.imshow(convertToRGB(test_image))
```

```python
def detect_faces(cascade, test_image, scaleFactor = 1.1):
    # create a copy of the image to prevent any changes to the original one.
    image_copy = test_image.copy()

    #convert the test image to gray scale as opencv face detector expects gray images
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

    # Applying the haar classifier to detect faces
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=5)

    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)

    return image_copy
```

```python
test_image2 = cv2.imread('test_2.jpg')

test_image_2_gray = cv2.cvtColor(test_image2, cv2.COLOR_BGR2GRAY)

plt.imshow(test_image_2_gray, cmap='gray')
   

faces_2 = detect_faces(haar_cascade_face, test_image2)

plt.imshow(convertToRGB(faces_2))
    
```

```python
test_image3 = cv2.imread('test_3.jpg')
test_image_3_gray = cv2.cvtColor(test_image3, cv2.COLOR_BGR2GRAY)
plt.imshow(test_image_3_gray, cmap='gray')

#call the function to detect faces
faces_3 = detect_faces(haar_cascade_face, test_image3)

#convert to RGB and display image
plt.imshow(convertToRGB(faces_3))#call the function to detect faces
  

```
